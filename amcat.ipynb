**Analyzing the Target Variable i.e Salary**

```python
# Histogram for Salary
plt.figure(figsize=(12,6))
df['Salary'].plot(kind='hist', bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.axvline(df['Salary'].mean(), color='red', linestyle='--', label='Mean')
plt.axvline(df['Salary'].median(), color='green', linestyle='--', label='Median')
plt.legend()
plt.show()
```

```python
# Boxplot for Salary
plt.figure(figsize=(8,5))
df['Salary'].plot(kind='box')
plt.title('Boxplot of Salary')
plt.xlabel('Salary')
plt.show()
```

```python
salary_95th_percentile = df['Salary'].quantile(0.95)
print(salary_95th_percentile)
```

```python
salary_99th_percentile = df['Salary'].quantile(0.99)
print(salary_99th_percentile)
```

```python
# Capping the outliers at 99th percentile
salary_99th_percentile = df['Salary'].quantile(0.99)
print(salary_99th_percentile)

df['Salary'] = df['Salary'].apply(lambda x: salary_99th_percentile if x > salary_99th_percentile else x)

df['Salary'].describe
```

```python
# Overview of other Numerical variables

# List of numerical columns to analyze
numerical_columns = ['10percentage', '12percentage', 'CollegeGPA', 'English', 'Logical', 'Quant']

# Plot histograms and boxplots
plt.figure(figsize=(18, 12))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(3, 2, i)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Overview of Categorical variables

# List of categorical columns to analyze
categorical_columns = ['CollegeTier', 'Degree', 'Designation', 'Gender']

# Count plots for categorical variables
plt.figure(figsize=(18, 12))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(2, 2, i)
    sns.countplot(y=df[col], order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)

plt.tight_layout()
plt.show()
```

**Bivariate Analysis**
Discover the relationships between numerical columns using Scatter plots, hexbin plots, pair plots, etc..
Identify the patterns between categorical and numerical columns using swarmplot, boxplot, barplot, etc..
Identify relationships between categorical and categorical columns using stacked bar plots.

```python
# Numerical(salary) vs Numerical
# List of numerical variables to analyze
numerical_vars = ['10percentage', '12percentage', 'CollegeGPA', 'English', 'Logical', 'Quant']

# Set the size of the plots
plt.figure(figsize=(15, 10))

# Create scatter plots
for i, var in enumerate(numerical_vars):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(data=df, x=var, y='Salary')
    plt.title(f'Salary vs {var}')
    plt.xlabel(var)
    plt.ylabel('Salary')

plt.tight_layout()
plt.show()

# Calculate and display the correlation coefficients
correlations = df[numerical_vars + ['Salary']].corr()['Salary'].drop('Salary')
print(correlations)
```

```python
# Numerical(i.e Salary) vs Categorical
# List of categorical variables to analyze
categorical_vars = ['CollegeTier', 'Degree', 'Designation', 'Gender']

# Set the size of the plots
plt.figure(figsize=(20, 12))

# Create box plots
for i, var in enumerate(categorical_vars):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, x=var, y='Salary')
    plt.title(f'Salary by {var}')
    plt.xlabel(var)
    plt.ylabel('Salary')

plt.tight_layout()
plt.show()
```

**Research Questions**
Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.
Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
# List unique designations
unique_designations = df['Designation'].unique()
print("Unique Designations in the dataset:")
print(unique_designations)
```

```python
# Normalize the Designation column
df['Designation'] = df['Designation'].str.strip().str.lower()

# Now check for unique designations again
unique_designations_normalized = df['Designation'].unique()
print("Normalized Unique Designations:")
print(unique_designations_normalized)
```

```python
# Designations to check
designations_to_check = ['programmer analyst', 'software engineer', 
                         'hardware engineer', 'associate engineer']

# Filter the data for these designations
salary_data = df[df['Designation'].str.lower().isin(designations_to_check)]

# Calculate average salary
average_salary = salary_data['Salary'].mean()
salary_range = salary_data['Salary'].describe()

# Display results
print(f"Average Salary: {average_salary}")
print(f"Salary Range:\n{salary_range}")
```

For second research question about gender and specialisation

```python
# Create a cross-tabulation of Gender and Specialization
gender_specialization_crosstab = pd.crosstab(df['Gender'], df['Specialization'])

# Display the cross-tabulation
print(gender_specialization_crosstab)
```

```python
# Create a stacked bar plot with a color palette
plt.figure(figsize=(14, 8))
colors = sns.color_palette('Spectral', n_colors=gender_specialization_crosstab.shape[1])  

# Plotting the data
gender_specialization_crosstab.plot(kind='bar', stacked=True, color=colors, alpha=0.8, ax=plt.gca())

# Set title and labels
plt.title('Gender Distribution by Specialization', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--')

# Add legend with designations
plt.legend(title='Specialization', fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

**Other Research Questions**

Analyzing the salary discrepancies by gender in specific designations.

```python
df['Salary'] = df['Salary'].astype(float)

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Create a box plot to show salary distribution by gender and designation
plt.figure(figsize=(16, 10))
sns.boxplot(data=df, x='Designation', y='Salary', hue='Gender', palette='Set2')

# Rotate x labels for better readability
plt.xticks(rotation=45, ha='right')

# Customize the plot
plt.title('Salary Discrepancies by Gender Across Designations')
plt.xlabel('Designation')
plt.ylabel('Salary (in Lakhs)')
plt.legend(title='Gender')
plt.tight_layout()

# Show the plot
plt.show()
```

Is there a significant difference in salary based on the specialization of the graduates?

```python
# Calculate average salary for each designation directly
df['Salary'] = df['Salary'].astype(float)  # Ensure the Salary column is float type
average_salary_all = df.groupby('Designation')['Salary'].mean()

# Sort the average salary for better visualization
average_salary_all = average_salary_all.sort_values(ascending=False)

# Create a bar plot for average salaries
plt.figure(figsize=(12, 8))
sns.barplot(x=average_salary_all.values, y=average_salary_all.index, palette='Set2')

# Customize the plot
plt.title('Average Salary by Designation')
plt.xlabel('Average Salary (in Lakhs)')
plt.ylabel('Designation')
plt.tight_layout()

# Show the plot
plt.show()
```
