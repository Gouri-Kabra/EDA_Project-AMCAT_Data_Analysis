### Analyzing the Target Variable i.e Salary

```python
# Non visual analysis 
min_salary = df['Salary'].min()
max_salary = df['Salary'].max()
mean_salary = df['Salary'].mean()
median_salary = df['Salary'].median()
range_salary = max_salary - min_salary
iqr_salary = df['Salary'].quantile(0.75) - df['Salary'].quantile(0.25)
std_salary = df['Salary'].std()
skew_salary = df['Salary'].skew()
kurtosis_salary = df['Salary'].kurt()

# Print the results (optional)
print("Non-visual Analysis of target variable i.e Salary:")
print(f"Min: {min_salary}")
print(f"Max: {max_salary}")
print(f"Mean: {mean_salary}")
print(f"Median: {median_salary}")
print(f"Range: {range_salary}")
print(f"IQR: {iqr_salary}")
print(f"Standard Deviation: {std_salary}")
print(f"Skewness: {skew_salary}")
print(f"Kurtosis: {kurtosis_salary}")
```

```python
# Set the style
sns.set(style="whitegrid")

# Create the figure
plt.figure(figsize=(12, 6))

# Plotting the histogram using seaborn with KDE
sns.histplot(df['Salary'], bins=30, kde=True, color='lightgreen', edgecolor='black', alpha=0.7)

# Add Mean and Median lines
plt.axvline(df['Salary'].mean(), color='orange', linestyle='--', label='Mean')  
plt.axvline(df['Salary'].median(), color='purple', linestyle='--', label='Median')  

# Titles and labels
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

**Salary Distribution**
- The histogram shows a right-skewed (positive skew) distribution, indicating that most of the salary data points are concentrated in lower range, with a long tail on the right. 
- The mean is slightly to the right of median, further confirming the right-skewed nature of the data.
- This suggests that majority of the graduates and post graduates earn below-average salaries, while a few individuals with very high salaries pull the mean higher.

```python
# Box plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Salary'], color='skyblue')
plt.title('Boxplot of Salary')
plt.xlabel('Salary', fontsize=15)
plt.xticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Analysis of Other Numerical Columns

```python
# other numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
print(numerical_cols)
```

```python
# List of numerical columns for analysis
numerical_columns = ['10percentage', '12percentage', 'CollegeGPA', 'English', 'Logical', 'Quant']

# Create a figure for distribution and box plots
fig, axes = plt.subplots(nrows=len(numerical_columns), ncols=2, figsize=(16, 5 * len(numerical_columns)))

for i, col in enumerate(numerical_columns):
    # Histogram with KDE
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i, 0], color='lightgreen', edgecolor='black', alpha=0.7)
    axes[i, 0].axvline(df[col].mean(), color='red', linestyle='--', label='Mean')
    axes[i, 0].axvline(df[col].median(), color='green', linestyle='--', label='Median')
    axes[i, 0].set_title(f'{col} Distribution', fontsize=16)
    axes[i, 0].set_xlabel(col, fontsize=12)
    axes[i, 0].set_ylabel('Frequency', fontsize=12)
    axes[i, 0].legend()

    # Boxplot
    sns.boxplot(x=df[col], ax=axes[i, 1], color='skyblue', fliersize=5)
    axes[i, 1].set_title(f'Boxplot of {col}', fontsize=16)
    axes[i, 1].set_xlabel(col, fontsize=12)

    # Identifying outliers
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f'Number of outliers in {col}: {len(outliers)}')

# Adjust layout
plt.tight_layout()
plt.show()
```

**Analysis of other Numerical Variables**
- Distributions (left Panels): All the variables - 10th percentage, 12th percentage, College GPA, English, Logical, and Quant scores --> appear to be close to a normal distribution, but some have slight skewness.
    - The 10th and 12th percentages, as well as the GPA, show a nearly symmetric distribution.
    - The English, Logical and Quant scores have slight peaks in the middle, suggesting that most scores hover around mean.
- Boxplots (Right Panels): 
    - The boxplots of 10th, 12th and GPA percentages show a few outliers, but generally the data is symmetrically distributed within the interquartile range.
    - The English, logical and Quant Scores have several outliers on both sides, indicating variability in the performance of the test takers

-- Comparison: The distributions and boxplots suggest that students generally perform within a narrow range for academic scores, while their logical and quantitative scores are more spread out.

### Analysing categorical columns

```python
# Categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print(categorical_cols)
```

```python
# List of categorical columns for analysis
categorical_columns = ['Designation', 'JobCity', 'Gender', 'Degree', 'Specialization', 'CollegeState']

N = 10  # Specify how many top categories to display

# Create a figure for subplots
num_columns = len(categorical_columns)
fig, axes = plt.subplots(nrows=(num_columns + 1) // 2, ncols=2, figsize=(15, 7 * ((num_columns + 1) // 2)))

# Flatten the axes array for easy iteration
axes = axes.flatten()

for i, col in enumerate(categorical_columns):
    top_categories = df[col].value_counts().nlargest(N)
    
    # Create bar plot
    top_categories.plot(kind='bar', ax=axes[i], color='lightblue', edgecolor='black')
    axes[i].set_title(f'Distribution of {col}', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Count', fontsize=12)
    axes[i].tick_params(axis='x', rotation=90)  # Set x-axis labels to vertical (90 degrees)

    # Add description as a text box
    textstr = '\n'.join([f'{cat}: {count}' for cat, count in zip(top_categories.index, top_categories.values)])
    
    # Create a textbox in each subplot
    props = dict(boxstyle='round', facecolor='white', alpha=0.5)
    axes[i].text(0.95, 0.95, textstr, transform=axes[i].transAxes, fontsize=10,
                 verticalalignment='top', horizontalalignment='right', bbox=props)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

**Insights from above categorical columns graphs:**
1. Analysing Designations: The most frequent job designations are Software Engineer, Software Developer, and System Engineer; with Software Engineer being the most common(539 instances). Specialized roles like Java Software Engineer and Senior Software Engineer also appear, but with lower frequencies(72-111 instances)
2. Analysing Job Cities: Bangalore is the top city for jobs, with 627 instances. Other top cities for jobs are Noida, Pune, Hyderabad, and Chennai also show significant job opportunities.
3. Analysing Gender: The data is skewed towards males with 3041 instances compared to 957 females. 
4. Analysing Degree: Most candidates hold a B.Tech/B.E degree(3700 instances), followed by MCA(243) and M.Tech/M.E (523). The frequency of M.Sc is much lower, with only 2 occurences.
5. Analysing Specialization: Popular specializations include Electronics and communication engineering(880) and computer science engineering(744). Specializations like mechanical engineering, electrical engineering also appear but with fewer entries.
6. Analysing College State: Uttarpradesh has the highest count of students(915), followed by Karnataka, Tamil Nadu and Telangana. Other states like Andhra Pradesh, Maharashtra and West Bengal also have notable presence.

*Analyzing the salary discrepancies by gender in specific designations.*

```python
# Identify the Top 10 Designations
top_designations = df['Designation'].value_counts().nlargest(10).index

# Filter the Data for Top 10 Designations
top_designations_data = df[df['Designation'].isin(top_designations)]

# Calculate Salary Statistics by Gender and Designation
salary_stats = top_designations_data.groupby(['Designation', 'Gender'])['Salary'].agg(['mean', 'median', 'count']).reset_index()
print(salary_stats)

# Visualize the Salary Discrepancies
plt.figure(figsize=(12, 6))
sns.barplot(data=salary_stats, x='Designation', y='mean', hue='Gender', palette='muted')
plt.title('Mean Salary by Gender in Top 10 Designations')
plt.xlabel('Designation')
plt.ylabel('Mean Salary')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()

# Boxplot for detailed salary distribution
plt.figure(figsize=(12, 6))
sns.boxplot(data=top_designations_data, x='Designation', y='Salary', hue='Gender', palette='muted', dodge=True)
plt.title('Salary Distribution by Gender in Top 10 Designations')
plt.xlabel('Designation')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

*How have job designations for graduates evolved over the past five years?*

```python
# extract the year
df['DOJ_year'] = df['DOJ'].dt.year  # Create a new column for the year of joining

# Get unique years available in the dataset
unique_years = df['DOJ_year'].unique()

# Sort the years and take the last 5 years
last_five_years = sorted(unique_years)[-5:]  # Get the last 5 unique years
recent_hires = df[df['DOJ_year'].isin(last_five_years)]

# Count the occurrences of each designation in the last five years
designation_counts = recent_hires['Designation'].value_counts()

# Get the top 10 designations
top_10_designations = designation_counts.nlargest(10).index.tolist()

# Filter the recent hires to include only top 10 designations
filtered_recent_hires = recent_hires[recent_hires['Designation'].isin(top_10_designations)]

# Count the occurrences of each top designation per joining year
designation_trends = filtered_recent_hires.groupby(['DOJ_year', 'Designation']).size().unstack(fill_value=0)

# Plotting the trends
plt.figure(figsize=(12, 6))
for designation in designation_trends.columns:
    plt.plot(designation_trends.index, designation_trends[designation], marker='o', label=designation)

plt.title('Trends of Top 10 Job Designations for Graduates Over the Last Five Years')
plt.xlabel('Year of Joining')
plt.ylabel('Number of Graduates')
plt.xticks(designation_trends.index)  # Set x-axis ticks to be the unique joining years
plt.legend(title='Designation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

The graph shows a growing trend towards technical roles among graduates, with Java-related positions being particularly popular. This suggests a positive outlook for career opportunities in the technology industry. Graduates who acquire proficiency in programming, software development, and related skills are likely to have better job prospects and career growth.
